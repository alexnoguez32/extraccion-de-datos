**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#filtro por objeto
filtro1=df[df["continent"]=="europe"]
filtro1
```

```python
#filtro por filas
#filtro2= df.iloc[300:302,:]    filas consecutivas
filtro2=df.iloc[[10, 20, 30, 40, 50] , : ] #filas no consecutivas
filtro2
```

```python
#filtro por columnas
filtro3=df.iloc[ : , 3:11]
#filtro3=df.iloc[ : , [1,4,6]] #columnas no consecutivas
filtro3
```

```python
df = pd.read_csv('CountryTable.csv',index_col=2)
df
```

```python
#paso 2: aplicamos filtro de filas vs columnas
filtro4=df.loc[["Europe","Africa"], ["local_name"]]
filtro4
```

```python
#filtro por cabecera
filtro5=df.head(15)
filtro5
```

```python
#filtro por comparacion <>
filtro6=df[df["population"] < 110000]
filtro6
```

```python
#FILTRO Y &
filtro7=df[(df["population"] < 110000) & (df["region"] == "Caribbean")]
filtro7
```

```python
#Filtro o
filtro8=df[(df["population"]< 110000) | (df["region"] == "Caribbean")]
filtro8
```

```python
#Filtro not   ~
filtro9=df[~(df["population"]<= 110000) & ~ (df["region"] == "Caribbean")]
filtro9
```
